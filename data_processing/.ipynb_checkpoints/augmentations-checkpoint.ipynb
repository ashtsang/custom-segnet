```python
'''
this script will generate augmentations based on tiled images and corresponding labels
- will apply augmentations to images/labels and curate train/test datasets
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import albumentations as A
```

```python
TILE_VERSION = 1.2
AUG_VERSION = 1.2
N_AUGS = 1
```

```python
## path to directories

tiled_data_path = r'\\babyserverdw3\PW Cloud Exp Documents\Lab work documenting\W-22-09-10 AT Build Competent multi task DL model for tissue labeling\dataset\1115\tiled\736x736_v{}'.format(TILE_VERSION)
save_path = r'\\babyserverdw3\PW Cloud Exp Documents\Lab work documenting\W-22-09-10 AT Build Competent multi task DL model for tissue labeling\dataset\1115\aug\aug_736x736_v{}'.format(AUG_VERSION)

if not os.path.exists(save_path):
    os.mkdir(save_path)
```

```python
## apply augmentations - affine, blur, rotation, color

def apply_augmentations(im, lb, mode_save_path, im_name):
    img = cv2.imread(im)
    label = cv2.imread(lb, cv2.IMREAD_UNCHANGED)
    # save original image and mask
    aug_counter = 0
#     fn = im_name + '_{}'.format(aug_counter)
#     cv2.imwrite(mode_save_path + '/images/' + fn + '.jpg', img)
#     cv2.imwrite(mode_save_path + '/labels/'+ fn + '.tif', label)
    transform = A.Compose([
        A.RandomBrightnessContrast(p=0.5),
        A.HueSaturationValue(hue_shift_limit=(-8, 8), sat_shift_limit=(-0.2, 0.2), p=0.5),
        ])
    aug_counter += 1
    # apply n augmentations per image
    for i in range(N_AUGS):
        transformed = transform(image=img, mask=label)
        fn = im_name + '_{}'.format(aug_counter)
        aug_counter += 1
        cv2.imwrite(mode_save_path + '/images/' + fn + '.tif', transformed['image'])
        cv2.imwrite(mode_save_path + '/labels/'+ fn + '.tif', transformed['mask'])
    return None
```

```python
## apply random augmentations to tiled images

for mode in os.listdir(tiled_data_path): # 'train', 'test'
    if mode == 'test':
        continue
    mode_save_path = os.path.join(save_path, mode)
    if not os.path.exists(mode_save_path): # make train/, test/ folders
        os.mkdir(mode_save_path)      
        os.mkdir(mode_save_path + '/images') # make images/, labels/ folders
        os.mkdir(mode_save_path + '/labels')
    images_path = os.path.join(tiled_data_path, mode + '/images')
    labels_path = os.path.join(tiled_data_path, mode + '/labels')    
    for fn in os.listdir(images_path): # for each image
        im_name = fn[:-4]
        im = os.path.join(images_path, im_name + '.tif') 
        lb = os.path.join(labels_path, im_name + '.tif')
        print('processing {}...'.format(im_name))
        apply_augmentations(im, lb, mode_save_path, im_name)
```
